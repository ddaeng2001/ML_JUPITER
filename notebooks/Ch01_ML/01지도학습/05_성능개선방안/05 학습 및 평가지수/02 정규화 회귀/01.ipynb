```python
# --------------------
# ğŸ“¦ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ë°ì´í„° ì¤€ë¹„
# --------------------
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge # Ridge íšŒê·€ ëª¨ë¸
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ì˜ˆì‹œ ë°ì´í„° ìƒì„± (íŠ¹ì„± 10ê°œ, ìƒ˜í”Œ 50ê°œ)
rng = np.random.RandomState(42)
X = rng.rand(50, 10) * 10
# íƒ€ê²Ÿê°’ ìƒì„± (ì„ í˜• ê´€ê³„ + ë…¸ì´ì¦ˆ)
y = 3 * X[:, 0] - 2 * X[:, 5] + 1 + rng.randn(50) * 5 

# ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("--- í•™ìŠµ ë°ì´í„° í¬ê¸° ---")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
```

```python
# --------------------
# ğŸ“ 2ë‹¨ê³„: ë°ì´í„° ìŠ¤ì¼€ì¼ë§ (í‘œì¤€í™”)
# --------------------
scaler = StandardScaler()
# í•™ìŠµ ë°ì´í„°ì— fit í›„ transform
X_train_scaled = scaler.fit_transform(X_train)
# í…ŒìŠ¤íŠ¸ ë°ì´í„°ëŠ” fit ì—†ì´ transformë§Œ
X_test_scaled = scaler.transform(X_test)

print("\n--- 2ë‹¨ê³„: ìŠ¤ì¼€ì¼ë§ ì™„ë£Œ ---")
```

```python
# --------------------
# ğŸ›¡ï¸ 3ë‹¨ê³„: Ridge ëª¨ë¸ í•™ìŠµ ë° ê³„ìˆ˜ í™•ì¸
# --------------------
# alpha=1.0: ê·œì œ ê°•ë„ (Î») ì„¤ì •. ê°’ì´ í´ìˆ˜ë¡ ê·œì œê°€ ê°•í•´ì§.
ridge_model = Ridge(alpha=1.0, random_state=42) 

# ëª¨ë¸ í•™ìŠµ
ridge_model.fit(X_train_scaled, y_train)

# í•™ìŠµëœ íšŒê·€ ê³„ìˆ˜ (ê°€ì¤‘ì¹˜) í™•ì¸
coefficients = ridge_model.coef_

print("\n--- 3ë‹¨ê³„: í•™ìŠµëœ íšŒê·€ ê³„ìˆ˜ (Î²) ---")
# RidgeëŠ” ê³„ìˆ˜ë¥¼ 0ì— ê°€ê¹ê²Œ ì¶•ì†Œí•˜ì§€ë§Œ, ì™„ì „íˆ 0ìœ¼ë¡œ ë§Œë“¤ì§€ëŠ” ì•ŠìŠµë‹ˆë‹¤.
print(coefficients)
```

```python
# --------------------
# ğŸ“‰ 4ë‹¨ê³„: ì˜ˆì¸¡ ë° ì„±ëŠ¥ í‰ê°€
# --------------------
from sklearn.metrics import mean_squared_error

# ì˜ˆì¸¡ ìˆ˜í–‰
y_pred = ridge_model.predict(X_test_scaled)

# ì„±ëŠ¥ í‰ê°€ (MSE ì‚¬ìš©)
mse = mean_squared_error(y_test, y_pred)

print(f"\n--- 4ë‹¨ê³„: ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ ---")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„° MSE (í‰ê·  ì œê³± ì˜¤ì°¨): {mse:.4f}")
```
