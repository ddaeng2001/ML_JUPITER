```python
# C-1) 스칼라(고정값)로 새 컬럼 추가
df["지점"] = "본점"

# # C-2) 계산식으로 파생 컬럼 추가 (할인가)
discount = 0.2
df["할인가"] = (df["가격"] * (1 - discount)).astype(int)

# # C-3) 조건 기반 컬럼(np.where / loc)
import numpy as np
df["고칼로리여부"] = np.where(df["칼로리"] >= 150, True, False)
df.loc[df["칼로리"] >= 150, "칼로리등급"] = "HIGH"
df.loc[df["칼로리"] < 150, "칼로리등급"] = "LOW"
df

# # C-4) 매핑/치환 기반 컬럼 생성(map/replace)
origin_map = {"아메리카노": "블렌드", "에스프레소": "싱글", "밀크티": "브렌드", "녹차": "티"}
df["원두타입"] = df["메뉴"].map(origin_map).fillna("기타")
df

# # C-5) 위치 지정해 컬럼 삽입(insert: 특정 위치에)
# df.insert(1, "메뉴길이", df["메뉴"].str.len())

# # C-6) assign로 체이닝
# df2 = (df
#        .assign(세전가격=lambda x: (x["가격"] / 1.1).round(0).astype(int),
#                이름대문자=lambda x: x["메뉴"].str.upper())
# )
```
